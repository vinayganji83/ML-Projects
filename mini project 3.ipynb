```python
# Categorical columns
categorical_cols = [
    "cp", "fbs", "restecg",
    "exang", "slope", "ca", "thal"
]

# Numerical columns
numerical_cols = [
    "age", "trestbps", "chol",
    "thalach", "oldpeak", "sex"
]
```

## encoding

```python
encoded_df=pd.get_dummies(df,columns=categorical_cols,drop_first=True)
```

```python
print("Original Shape:",df.shape)
print("Encoded Shape:",encoded_df.shape)
```
